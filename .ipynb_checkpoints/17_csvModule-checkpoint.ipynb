```python
with open("NIFTY 200_Historical_PR_01072025to01072026.csv","r") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    # next(csv_reader)
    data = list(csv_reader)
    print(data[0].keys())
    print(len(data))
   
    window = 7

    for i in range(len(data)-window+1):
     moving_average_window = sum([float(data[i+j]["Close"])  for j in range(window)])/window
     data[i]["MA7"] = round(moving_average_window,2)
print(data)
        
   
```

```python
field_names = data[0].keys()
```

```python
with open("NIFTY 200_Historical_PR_01072025to0107202_copy.csv","w") as new_csv:
    
    csv_writer = csv.DictWriter(new_csv,fieldnames=field_names)
    csv_writer.writerows(data)
   
    
```
