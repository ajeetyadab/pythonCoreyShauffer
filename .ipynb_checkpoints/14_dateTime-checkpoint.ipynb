#### calculating age

```python
tday = datetime.date.today()
```

```python
age = tday- birth_date
print(age)
```

```python
start_year = 1988
current_year = datetime.datetime.today().year
```

```python
leap_year_since_your_birthdays = []
for i in range(start_year,current_year+1):
    if( i % 4 ==0 and i % 100 !=0) or ( i % 400 ==0):
        leap_year_since_your_birthdays.append(i)
```

```python
leap_year_since_your_birthdays
```
