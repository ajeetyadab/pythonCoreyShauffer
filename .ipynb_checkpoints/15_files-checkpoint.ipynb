```python
with open("notes.txt","r") as f:
    size_to_read = 5
    f_contents = f.read(size_to_read)
    print(f_contents)
```

```python
with open("notes.txt","r+") as f:
    # f.write("line6")
    print(f.read())
```

```python
with open("notes.txt","a+") as f:
   
    f.write("line 7 \n")
    print("again")
    f.seek(0)
    print(f.read())
```

```python
with open("notes.txt","r") as rf:
    with open("notes.txt","w") as wf:
        for line in rf:
            wf.write(line)
```
